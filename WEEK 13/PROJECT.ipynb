```python
import pandas as pd


df1 = pd.DataFrame({"Student Name" : [],
                   "Matric Number": [],
                   "Department": [],
                   "Level":[]})

for i in range(5):
    name = input("Enter The Student Name: ")
    matric_num = input("Enter The Matric Number: ")
    department = input("Enter The Department: ")
    level = input("Enter The Level: ")
    
    df2 =   pd.DataFrame({"Student Name" : [name],
                          "Matric Number": [matric_num],
                          "Department": [department],
                          "Level":[level]})
    df1 = pd.concat([df1, df2], ignore_index = True)
    df1.reset_index()
    
df1.to_csv("PAU SIMS.csv")
df1
```

```python
import pandas as pd

assests = [1000000, 1500000, 3000000, 250000, 800000]
liabilities = [200000, 500000, 1500000, 50000, 300000]
percentage_leverage_list = []

for i in range(5):
    percentage_leverage = (assests[i] - liabilities[i]) / (assests[i]) * 100
    percentage_leverage = round(percentage_leverage)
    percentage_leverage_list.append(percentage_leverage)
    
df =   pd.DataFrame({"Company" : ["Enron", "Anderson", "GK Jones", "Mica", "Dune"],
                     "Founded": ["1987", "1936", "2001", "1996", "2008"],
                     "Assests": [1000000, 1500000, 3000000, 250000, 800000],
                     "Liabilities": [200000, 500000, 1500000, 50000, 300000],
                     "Percentage Leverages (%)": [x for x in percentage_leverage_list]
                    })
df.to_csv("Company data.csv")
df
```
